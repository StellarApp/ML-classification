```python
import matplotlib.pyplot as plt
import numpy as np
import math
# %matplotlib inline
```

```python
x = np.linspace(-10, 10, 100, endpoint = True) # stop is the last sample
z = 1/(1 + np.exp(-x))
```

```python
plt.plot(x, z)
plt.xlabel("x")
plt.ylabel("Sigmoid(X)")
plt.show()
```
